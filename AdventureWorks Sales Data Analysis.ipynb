```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_data = pd.read_csv(r"C:\Users\SIVARAM\Downloads\AdventureWorks Raw Data\AdventureWorks Sales Data 2020.csv")
```

```python
sales_data['OrderDate'] = pd.to_datetime(sales_data['OrderDate'])
```

```python
sales_data['YearMonth'] = sales_data['OrderDate'].dt.to_period('M')
```

```python
monthly_sales = sales_data.groupby('YearMonth')['OrderQuantity'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].astype(str)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='YearMonth', y='OrderQuantity', marker='o')
plt.title('Total Order Quantity Over Time')
plt.xlabel('Date')
plt.ylabel('Total Order Quantity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
territory_sales = sales_data.groupby('TerritoryKey')['OrderQuantity'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=territory_sales, x='TerritoryKey', y='OrderQuantity', palette='viridis')
plt.title('Total Order Quantity by Territory')
plt.xlabel('Territory')
plt.ylabel('Total Order Quantity')
plt.tight_layout()
plt.show()
```
